In our pursuit of uncovering the intricacies of healthcare insurance costs, we shift our focus to evaluating the performance of a model. We will take 3 machine learning models namely Random Forest Regressor, Gradient Boost Regressor, and XGBoost Regressor and evaluate how well they perform by employing the cross-validation.

###### Assessing the Models.
- Import the numpy library as 'np' to perform numerical operations and the necessary libraries for machine learning tasks, including 'cross_val_score', 'RandomForestRegressor','GradientBoostRegressor',and 'XGBRegressor'.
- Create an instance of the 'RandomForestRegressor' class and assign it to 'rand_forest_model'. This model is a random forest regressor with specific configurations.
    - 'n_estimators=50' indicates that the random forest will consist of 50 decision trees.
    - 'n_jobs=2' specifies that two CPU cores should be used for parallel processing.
    - 'random_state=42' sets the random seed for reproducibility.
- Create an instance of the 'GradientBoostingRegressor' class and assign it to 'grad_boost_model'. This model is a gradient boosting regressor with the configurations.
    - 'random_state =42' which sets the random seed for reproducibility
- Create an instance of the 'XGBRegressor' class and assign it to 'xboost_model'. This model is an extreme gradient boosting regressor with the configurations.
    - 'random_state =42' which sets the random seed for reproducibility
- Use 'cross_val_score' to iteratively evaluate the performance of the models in a 'for' loop. It takes the following arguments:
    - 'model': List of machine learning models to be evaluated.
    - 'X_train': The training feature data.
    - 'y_train': The corresponding target values.
    - 'scoring="neg_root_mean_squared_error"': The scoring metric used to assess model performance, which is the negative root mean squared error.
    - 'cv=10': The number of cross-validation folds, which is set to 10. Cross-validation helps assess a model's generalization performance.
- The 'cross_val_score' function returns an array of negative root mean squared error scores (RMSE) for each fold of the cross-validation, which are more interpretable.
- Compute the standard deviation ('std') of the RMSE values, which provides information about the variability of model performance across different cross-validation folds. A lower standard deviation indicates more consistent model performance.
- Use 'cross_val_score' to find out the performance but this time with another scoring metric.
    - 'scoring = "r2": The scoring metric used to assess model performance, which is the R2 score.
- The 'cross_val_score' function returns an array of R2 scores (r2) for each fold of the cross-validation, which tells us about the goodness of the fit of the regression model.
- Compute the standard deviation ('std') of the r2 values, which provides information about the variability of model performance across different cross-validation folds. A lower standard deviation indicates more consistent model performance.

'std' represents how much the model's performance varies across the different folds of the cross-validation, giving insight into the model's stability and robustness.

```python

import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor

# creating instances for the models
rand_forest_model = RandomForestRegressor(n_estimators=50, n_jobs=2, random_state=42)
grad_boost_model = GradientBoostingRegressor(n_estimators=50,max_features='auto',random_state=42)
xboost_model = XGBRegressor(n_estimators=50, random_state=42)
```

```python
# validating the models using cross_val_score in a for loop

models = [rand_forest_model,grad_boost_model,xboost_model]
model_name = ['RandomForestRegressor','GradientBoosting Regressor','XGBoostRegressor']
compare_dataframe = pd.DataFrame(index=['rmse.std','rmse.mean','r2.std','r2.mean','score'])
for i, model in enumerate(models):
    rmse = cross_val_score(model,x_train,y_train,scoring='neg_root_mean_squared_error',cv=10)
    r2 = cross_val_score(model,x_train,y_train,scoring='r2',cv=10)
    scores = 0 #model.score(x_train,y_train)
    compare_dataframe.insert(loc=0,column=model_name[i],value=[rmse.std(),rmse.mean(), r2.std(),r2.mean(),scores])


# the dataframe makes it easier for us to compare the performances of the models we are evaluating
compare_dataframe
```

##### Training a Model, Making Predictions, and Comparing Results.
- A used defined function 'evaluate_model' is created to evaluate the model's performance. This is for reusability and readability of the code.
- The model is being trained using the training dataset, 'X_train' for features and 'y_train' for target values.
- After training, the model is ready to make predictions on new data.
- 'predictions' is created by using the trained model to predict target values for the 'X_test' dataset, which is a set of feature data from the testing dataset.
- To make the predicted values more readable, 'rounded_predictions' is created by rounding the 'predictions' to two decimal places using the 'np.round()' function.
- A comparison of the first 10 entries of the actual charges ('y_test' - it should be also rounded to 2 decimal places) and the predicted charges ('rounded_predictions') is created in a DataFrame called 'compare'.
- The 'compare' DataFrame displays both the actual and predicted charges for the first 10 instances, allowing for a quick visual assessment of how well the model's predictions align with the actual values.
- This type of comparison is valuable for understanding the model's accuracy on unseen data.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

def evaluate_model(s,y_true,y_predict):
    '''
    The function calculates the model metrics and prints them.
    the parameters are:
    s: shape of the dataframe
    y_true: actual outcome values
    y_predict: the predicted outcome values
    '''
    mse = mean_squared_error(y_true,y_predict)
    mae = mean_absolute_error(y_true,y_predict)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true,y_predict)
    P = s[1] #len(x_train.columns)--> no. of columns
    N = s[0] #len(x_train.index)--> no. of rows
    adjr2 = (1-((1-r2)*(N-1)/(N-P-1)))
    print("Mean absolute Error : {:0.4f}".format(mae))
    print("Mean Squared Error : {:0.4f}".format(mse))
    print("Root Mean Squared Error : {:0.4f}".format(rmse))
    print("R2 Score : {:0.4f}".format(r2))
    print("Adjusted R2 Score : {:0.4f}".format(adjr2))

```

```python

rand_forest_model.fit(x_train,y_train)

predictions = rand_forest_model.predict(x_test)

rounded_predictions = np.round(predictions, decimals=2)
y_test_rounded = np.round(y_test.values,decimals=2)
compare = {'actual': y_test_rounded[:10],'predictionrandforest': rounded_predictions[:10]}
compare = pd.DataFrame(compare)

evaluate_model(x_train.shape,y_test,predictions)
#--- Inspect data ---

compare
```

```python
grad_boost_model.fit(x_train,y_train)

predictions2 = grad_boost_model.predict(x_test)

rounded_predictions2 = np.round(predictions2, decimals=2)
y_test_rounded = np.round(y_test.values,decimals=2)
#compare = {'actual': y_test_rounded[:10],'prediction': rounded_predictions2[:10]}
compare['predictionsgradboost']=rounded_predictions2[:10]

evaluate_model(x_train.shape,y_test,predictions2)
#--- Inspect data ---

print(compare)
```
